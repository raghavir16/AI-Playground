```python
from autogen import UserProxyAgent, AssistantAgent, GroupChat, GroupChatManager
import os

# Set your OpenAI API key (can also use local models via config)
#os.environ["OPENAI_API_KEY"] = "openai-key"

llm_config = {
    "config_list": [{"model": "gpt-4", "api_key": "api_key"}],
    "timeout": 120,
}

requirements_analyst = AssistantAgent(
    name="RequirementsAnalyst",
    llm_config=llm_config,
    system_message="You are a business analyst. Extract clear and complete project requirements from the client's brief."
)

technical_architect = AssistantAgent(
    name="TechnicalArchitect",
    llm_config=llm_config,
    system_message="You are a solutions architect. Use the requirements to design a scalable technical architecture."
)

proposal_writer = AssistantAgent(
    name="ProposalWriter",
    llm_config=llm_config,
    system_message="You are a proposal writer. Write a detailed project proposal using input from the analyst and architect."
)

user_proxy = UserProxyAgent(
    name="ClientProxy",
    human_input_mode="NEVER",
    code_execution_config={"use_docker": False}
)

# Define group chat
groupchat = GroupChat(
    agents=[user_proxy, requirements_analyst, technical_architect, proposal_writer],
    messages=[],
    max_round=10
)

manager = GroupChatManager(groupchat=groupchat, llm_config=llm_config)

# Start conversation
client_brief = """Client: NovaCorp
Project: Cloud-native transformation of internal tools.
Goals: Migrate legacy Java apps to microservices, enable CI/CD, adopt Kubernetes, and improve cost efficiency.
Budget: $150,000
Timeline: 6 months"""

user_proxy.initiate_chat(
    manager,
    message=f"Team, here's the client brief. Please collaborate to produce a full project proposal:\n\n{client_brief}"
)
```
